```python
#The fire contained date(CONT_DATE) has many NaN values. Will replace the NaN values with the fire Discovery Date.

wildfire_df["CONT_DATE"] = wildfire_df["CONT_DATE"].fillna(wildfire_df['DISCOVERY_DATE'])

wildfire_df.loc[:, ["DISCOVERY_DATE", "CONT_DATE"]]

#wildfire_df[wildfire_df["CONT_DATE"] == None ]
#wildfire_df["CONT_DATE"] = wildfire_df[wildfire_df["CONT_DATE"].isna()]["DISCOVERY_DATE"]
#wildfire_df.loc[:, ["DISCOVERY_DATE", "CONT_DATE"]]
#wildfire_df[["CONT_DATE"]]
#df[filter boolean].loc[extract columns]
```

```python
import datetime
```

```python
#Convert the Julian date formats to regular date formats

wildfire_df["DISCOVERY_DATE_NEW"] = pd.to_datetime(wildfire_df["DISCOVERY_DATE"], unit="D", origin="julian")
wildfire_df["CONT_DATE_NEW"] = pd.to_datetime(wildfire_df["CONT_DATE"], unit="D", origin="julian")

wildfire_df.loc[:, ["DISCOVERY_DATE_NEW", "CONT_DATE_NEW"]]
```

```python
wildfire_reduced_df = wildfire_df.loc[:, ["OBJECTID", "SOURCE_SYSTEM_TYPE", "NWCG_REPORTING_UNIT_ID", 
                                          "NWCG_REPORTING_UNIT_NAME", "SOURCE_REPORTING_UNIT", 
                                          "SOURCE_REPORTING_UNIT_NAME", "FIRE_NAME", "FIRE_YEAR",
                                         "DISCOVERY_DATE", "DISCOVERY_DATE_NEW", "DISCOVERY_DOY", "DISCOVERY_TIME",
                                          "CONT_DATE", "CONT_DATE_NEW", "CONT_DOY", "CONT_TIME", "FIRE_SIZE",
                                          "FIRE_SIZE_CLASS", "LATITUDE", "LONGITUDE", "OWNER_CODE", "OWNER_DESCR",
                                         "STATE", "COUNTY", "FIPS_CODE", "FIPS_NAME"]]
wildfire_reduced_df.head()
```

```python
#fill in the NaN values with zeor and none

#wildfire_reduced_df.dtype

wildfire_reduced_df.loc[:, ["FIRE_NAME", "DISCOVERY_TIME", "CONT_TIME", 
                           "COUNTY", "FIPS_CODE", "FIPS_NAME"]].fillna("NA")
wildfire_reduced_df.loc[:, ["CONT_DOY", ]].fillna(0.0)
```
